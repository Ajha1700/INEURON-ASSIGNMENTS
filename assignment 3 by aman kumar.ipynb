```python
2. When does the code in a function run: when its specified or when its called?
    
ANSWER

In Python, the code in a function runs when the function is called, not when it is specified.
When we define a function, we are essentially creating a block of code that will be executed when the function is called.
The function definition itself does not execute the code it contains.

For example, consider the following Python code:

def print_hello():
    print("Hello, World!")
    
print("Before function call")
print_hello()
print("After function call")

In this code, the print_hello function is defined, but the code inside the function is not executed until the function
is called. When the program runs, it will print "Before function call" first, 
then call the print_hello function, which will print "Hello, World!". Finally, it will print "After function call".

So,the code in a function runs when the function is called, and not when it is defined.




```

```python
3. What statement creates a function?
ANSWER

To create a function in Python, you use the def statement. The def statement is followed by the name of the function, 
parentheses (), and a colon :. The body of the function is then indented below the def statement.
Here's an example:

def greet(name):
    print("Hello, " + name + "!")
    
    
In this example, we define a function called greet that takes one argument name. 
The body of the function is a single statement that prints a greeting message that includes the name argument.

To call the greet function, we would simply use its name followed by parentheses and the argument
we want to pass to the function, like this:

greet("Alice")


This would call the greet function with the argument "Alice", and the output would be Hello, Alice!.

So, to create a function in Python, you use the def statement followed by the function name, 
parentheses for the function arguments, and a colon.

```

```python
4. What is the difference between a function and a function call?
ANSWER

In Python, a function is a block of code that performs a specific task and can be reused throughout a program.
It is defined using the def keyword, followed by the function name, parentheses for any arguments, and a colon. 
The body of the function is then indented below the def statement.

On the other hand, a function call is the process of invoking or executing a function. 
To call a function in Python, we simply use the name of the function followed by parentheses with any necessary 
arguments inside the parentheses. When a function is called, the code inside the function body is executed.

For example, 

def greet(name):
    print("Hello, " + name + "!")
    
greet("Aman")

In this code, greet is a function that takes one argument name and prints a greeting message that includes the name argument. 
The function is defined using the def keyword.

The second line of the code is a function call, where we pass the argument "Aman" to the greet function.
This causes the code inside the greet function to execute, printing the message "Hello, Aman!" to the console.

So, a function is a reusable block of code that performs a specific task, while a function call is the process of executing
or invoking that function with the necessary arguments.
```

```python
5. How many global scopes are there in a Python program? How many local scopes?
ANSWER


In Python, there is only one global scope per program. 
This means that any variable that is defined outside of a function can be accessed from any part of the program.
Once a variable is defined in the global scope, it is available throughout the program, including inside any functions.

On the other hand, 
local scopes are created every time a function is called. This means that any variable that is defined 
inside a function is only accessible within that function.
When the function completes its execution, the local scope is destroyed and any variables defined within that scope
are also destroyed.

Its important to note that while there can be multiple local scopes in a Python program (one for each function call),
each local scope is separate and distinct from the others. Variables defined within one local scope cannot be accessed
from another local scope. Similarly, variables defined within a function cannot be accessed outside of that function, 
unless they are explicitly returned or made global.

So, there is only one global scope per Python program, while local scopes are created every time a function is called.



```

```python
6. What happens to variables in a local scope when the function call returns?
ANSWER

When a function call returns, the local scope that was created for that function is destroyed. 
Any variables that were defined within the local scope are also destroyed and are no longer accessible.

For example,

def calculate_sum(a, b):
    total = a + b
    return total

result = calculate_sum(2, 3)
print(result)

In this code, the calculate_sum function takes two arguments a and b and calculates their sum,
which is stored in a local variable called total.The function then returns the value of total.

When the function is called with arguments 2 and 3, a local scope is created, and the variables a, b, and 
total are defined within that scope. The sum of a and b is calculated and stored in total, and then total is returned
by the function. Finally, the local scope is destroyed and the variables a, b, and total are no longer accessible.

The returned value of total is assigned to the variable result, which is then printed to the console. This output will be 5.

so, when a function call returns in Python, the local scope that was created for that function
is destroyed, and any variables defined within that scope are also destroyed and are no longer accessible.



```

```python
7. What is the concept of a return value? Is it possible to have a return value in an expression?

ANSWER

In Python, a return value is the value that a function sends back to the calling code after it completes its execution.
A return value can be any Python object, such as a number, string, list, or even another function.

To specify a return value in a function, you use the return statement followed by the value or expression that 
you want to return.
For example, 
def square(x):
    return x * x

In this function, the return statement sends back the value of x * x to the calling code after it completes its execution.
If we call this function with an argument of 3, the function will return the value 9.

Yes, it is possible to have a return value in an expression. 
In Python, we can use a function call as an expression in any part of your code where an expression is allowed. 
This means that you can call a function and use its return value in the same line of code.

For example,
result = square(3) + 5
print(result)
In this code, the square function is called with an argument of 3, which returns the value 9. The returned value 
is then added to 5 to produce the final value of 14, which is stored in the variable result.
The value of result is then printed to the console.

so, a return value in Python is the value that a function sends back to the calling code after it completes its execution.
It can be any Python object, and it is possible to use a function call as an expression in any part of our code where
an expression is allowed.

```

```python
8. If a function does not have a return statement, what is the return value of a call to that function?

ANSWER

If a function does not have a return statement, it implicitly returns None, which is a special Python object
that represents the absence of a value. This means that a call to a function without a return statement will always return None.

For example, 

def print_message(message):
    print("The message is:", message)

This function takes a message as an argument and prints it to the console using the print() function. 
Since there is no return statement in this function, calling it will always return None. 
For example:

result = print_message("Hello, world!")
print(result)

In this code, the print_message function is called with an argument of "Hello, world!". The function prints the message to 
the console, but it does not return a value. The value of result is therefore None, which is assigned to the variable. 
When None is printed to the console, it produces the output None.

so, if a function does not have a return statement, it implicitly returns None, which is a special Python object
that represents the absence of a value. Calling a function without a return statement will always return None.



```

```python
9. How do you make a function variable refer to the global variable?
ANSWER

In Python, to make a function variable refer to a global variable, you can use the global keyword inside the function.
This tells Python that the variable being used is a global variable rather than a local variable.

Heres an example:
    
x = 10

def print_x():
    global x
    print(x)

print_x()

In this example, we define a global variable x with a value of 10. We then define a function called print_x that prints 
the value of x. Inside the function, we use the global keyword to indicate that we want to use the global variable x 
rather than a local variable.

When we call the print_x function, it will print the value of the global variable x, which is 10.

Its important to note that using global variables can make your code harder to read and understand, 
since its not always clear where a variable is defined and how its being used. In general, its best to use local variables
within functions whenever possible, and only use global variables when absolutely necessary.




```

```python
14. What is the purpose of the try clause? What is the purpose of the except clause?
ANSWER

In Python, the try and except clauses are used for error handling. The purpose of the try clause is to enclose the code 
that might raise an exception, and the purpose of the except clause is to define the code that will be executed if
an exception is raised.

try clause: The try clause is used to enclose the code that might raise an exception.
When an exception is raised inside the try clause, the program execution is immediately transferred to the except clause.

except clause: The except clause is used to define the code that will be executed if an exception is raised inside
the try clause. You can use one or more except clauses to catch specific exceptions or types of exceptions, or 
you can use a general except clause to catch any exception.
```
